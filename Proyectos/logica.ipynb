# Introducción a las compuertas lógicas

En lógica proposicional y en circuitos digitales, las **compuertas lógicas** permiten construir expresiones booleanas a partir de proposiciones. Cada compuerta tiene un símbolo lógico asociado que se representa comúnmente con notación matemática.

A continuación se presentan las principales compuertas:

- **AND (conjunción)**: solo es verdadera si ambas entradas lo son.  
  Símbolo: $p \wedge q$

- **OR (disyunción)**: es verdadera si al menos una entrada lo es.  
  Símbolo: $p \vee q$

- **NOT (negación)**: invierte el valor lógico de una proposición.  
  Símbolo: $\neg p$

- **IMPLICACIÓN**: la proposición $p$ implica a $q$.  
  Símbolo: $p \rightarrow q$

- **BICONDICIONAL**: ambas proposiciones son equivalentes.  
  Símbolo: $p \leftrightarrow q$

Estas operaciones permiten construir expresiones complejas como:  
$(p \wedge \neg q) \vee (r \rightarrow s)$

```python
def main():
    # Diccionario para almacenar los valores de verdad de las proposiciones
    valores_proposiciones = {}
    
    # Lista de posibles letras proposicionales
    letras_disponibles = ['p', 'q']
    
    print("===== EVALUADOR DE PROPOSICIONES LÓGICAS =====")
    print("Este programa te permite evaluar expresiones lógicas.")
    
    # Preguntar por los valores de cada letra proposicional
    for letra in letras_disponibles:
        usar = input(f"¿Deseas asignar un valor a la letra '{letra}'? (s/n): ").strip().lower()
        
        if usar == 's':
            while True:
                try:
                    valor = int(input(f"Ingresa el valor de verdad para '{letra}' (0 para Falso, 1 para Verdadero): "))
                    if valor not in [0, 1]:
                        print("Error: Solo se aceptan los valores 0 o 1.")
                        continue
                    valores_proposiciones[letra] = bool(valor)
                    break
                except ValueError:
                    print("Error: Debes ingresar un número (0 o 1).")
    
    # Mostrar las letras que tienen valores asignados
    print("\nValores asignados:")
    for letra, valor in valores_proposiciones.items():
        print(f"{letra}: {1 if valor else 0}")
    
    # Si no hay letras con valores asignados, terminar el programa
    if not valores_proposiciones:
        print("No has asignado valores a ninguna letra proposicional. No se puede evaluar ninguna expresión.")
        return
    
    # Solicitar la expresión lógica
    print("\nAhora puedes ingresar una expresión lógica utilizando las letras que has definido.")
    print("Operadores disponibles:")
    print("  'y': conjunción (AND)")
    print("  'o': disyunción (OR)")
    print("  'no': negación (NOT)")
    print("  '->': implicación")
    print("  '<->': doble implicación (bicondicional)")
    print("Ejemplo: 'p y q' o 'no p o q'")
    
    expresion = input("\nIngresa tu expresión: ").strip().lower()
    
    # Evaluar la expresión
    try:
        resultado = evaluar_expresion(expresion, valores_proposiciones)
        print(f"\nLa expresión '{expresion}' es {1 if resultado else 0} (", end="")
        print("Verdadero)" if resultado else "Falso)")
    except Exception as e:
        print(f"Error al evaluar la expresión: {e}")

def evaluar_expresion(expresion, valores):
    """Evalúa una expresión lógica según los valores de verdad asignados"""
    # Reemplazar palabras por operadores de Python
    expresion_python = expresion
    
    # Verificar que todas las letras en la expresión tienen valores asignados
    for letra in 'pqrst':
        if letra in expresion_python and letra not in valores:
            raise ValueError(f"La letra '{letra}' está en la expresión pero no tiene un valor asignado.")
    
    # Reemplazar letras por sus valores
    for letra, valor in valores.items():
        #expresion_python = expresion_python.replace(letra, f" {str(valor).lower()} ")
        expresion_python = expresion_python.replace(letra, f" {str(valor)} ")
        print("exp es ", expresion_python)
    
    # Reemplazar operadores lógicos por operadores de Python
    expresion_python = expresion_python.replace(" y ", " and ")
    expresion_python = expresion_python.replace(" o ", " or ")
    expresion_python = expresion_python.replace("no ", "not ")
    
    # Manejar implicaciones (p -> q equivale a not p or q)
    while "->" in expresion_python:
        inicio = expresion_python.find("->")
        
        # Encontrar el operando izquierdo
        contador = inicio - 1
        parentesis = 0
        while contador >= 0:
            if expresion_python[contador] == ")":
                parentesis += 1
            elif expresion_python[contador] == "(":
                parentesis -= 1
            
            if parentesis < 0 or (parentesis == 0 and expresion_python[contador] in [" ", "("] and contador > 0 and expresion_python[contador-1] not in ["and", "or", "not"]):
                break
            contador -= 1
        
        izquierdo = expresion_python[contador+1:inicio].strip()
        if not izquierdo.startswith("(") or not izquierdo.endswith(")"):
            izquierdo = f"({izquierdo})"
        
        # Encontrar el operando derecho
        contador = inicio + 2
        parentesis = 0
        while contador < len(expresion_python):
            if expresion_python[contador] == "(":
                parentesis += 1
            elif expresion_python[contador] == ")":
                parentesis -= 1
            
            if parentesis < 0 or (parentesis == 0 and expresion_python[contador] in [" ", ")"] and contador < len(expresion_python)-1 and expresion_python[contador+1] not in ["and", "or"]):
                break
            contador += 1
        
        derecho = expresion_python[inicio+2:contador+1].strip()
        if not derecho.startswith("(") or not derecho.endswith(")"):
            derecho = f"({derecho})"
        
        # Reemplazar p -> q con not p or q
        reemplazo = f"(not {izquierdo} or {derecho})"
        expresion_python = expresion_python[:contador+1].replace(f"{izquierdo} -> {derecho}", reemplazo) + expresion_python[contador+1:]
    
    # Manejar bicondicionales (p <-> q equivale a (p -> q) and (q -> p))
    while "<->" in expresion_python:
        inicio = expresion_python.find("<->")
        
        # Encontrar el operando izquierdo
        contador = inicio - 1
        parentesis = 0
        while contador >= 0:
            if expresion_python[contador] == ")":
                parentesis += 1
            elif expresion_python[contador] == "(":
                parentesis -= 1
            
            if parentesis < 0 or (parentesis == 0 and expresion_python[contador] in [" ", "("] and contador > 0 and expresion_python[contador-1] not in ["and", "or", "not"]):
                break
            contador -= 1
        
        izquierdo = expresion_python[contador+1:inicio].strip()
        if not izquierdo.startswith("(") or not izquierdo.endswith(")"):
            izquierdo = f"({izquierdo})"
        
        # Encontrar el operando derecho
        contador = inicio + 3
        parentesis = 0
        while contador < len(expresion_python):
            if expresion_python[contador] == "(":
                parentesis += 1
            elif expresion_python[contador] == ")":
                parentesis -= 1
            
            if parentesis < 0 or (parentesis == 0 and expresion_python[contador] in [" ", ")"] and contador < len(expresion_python)-1 and expresion_python[contador+1] not in ["and", "or"]):
                break
            contador += 1
        
        derecho = expresion_python[inicio+3:contador+1].strip()
        if not derecho.startswith("(") or not derecho.endswith(")"):
            derecho = f"({derecho})"
        
        # Reemplazar p <-> q con (not p or q) and (not q or p)
        reemplazo = f"((not {izquierdo} or {derecho}) and (not {derecho} or {izquierdo}))"
        expresion_python = expresion_python[:contador+1].replace(f"{izquierdo} <-> {derecho}", reemplazo) + expresion_python[contador+1:]
    
    # Evaluar la expresión Python
    try:
        print(expresion_python)
        resultado = eval(expresion_python)
        return resultado
    except Exception as e:
        raise ValueError(f"No se pudo evaluar la expresión: {str(e)}. Expresión traducida: {expresion_python}")

if __name__ == "__main__":
    main()
```
