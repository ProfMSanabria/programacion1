## a. Cruce de Decisiones: Condicionales

Las estructuras de control son como los signos de tr치fico en el camino de la ejecuci칩n de nuestro programa. Nos permiten tomar decisiones basadas en diferentes condiciones y repetir ciertas acciones hasta que se cumpla una condici칩n. Vamos a imaginar que estamos desarrollando un programa para determinar si un estudiante ha aprobado o reprobado un examen basado en su calificaci칩n.

```python
calificacion = int(input("Por favor ingrese la calificaci칩n del estudiante: "))

if calificacion >= 6:
    print("El estudiante ha aprobado el examen.")
else:
    print("El estudiante ha reprobado el examen.")
```

En este ejemplo, el programa pide al usuario que ingrese la calificaci칩n del estudiante. Luego, usa una declaraci칩n if para comprobar si la calificaci칩n es igual o superior a 6. A menudo, cuando programamos, nos encontramos con situaciones donde no basta con preguntar "쯘s esto cierto o falso?". La realidad suele ser m치s compleja y necesitamos considerar m칰ltiples condiciones. Por ejemplo, 쯤u칠 pasa si queremos saber si un n칰mero es positivo, negativo o cero? No podemos limitarnos a una sola pregunta. Aqu칤 es donde el uso de elif se vuelve esencial, permiti칠ndonos manejar varios escenarios posibles con claridad y precisi칩n.

```python
numero = float(input("Ingrese un n칰mero: "))

if numero > 0:
    print(f"{numero} es un n칰mero positivo.")
elif numero == 0:
    print(f"{numero} es cero.")
else:
    print(f"{numero} es un n칰mero negativo.")
```

```python
numero = int(input("Ingrese un n칰mero: "))
tipo_numero = "par" if numero % 2 == 0 else "impar"
print(f"El n칰mero {numero} es {tipo_numero}.")
```

## c. Please, stop the 'while'

Primero vamos a ver que es "while", es un bucle que repite un bloque de codigo mientras sea verdadera, en nuestras palabras imaginemos que tenemos 10 ml de agua este estara con agua que es nuestra verdad hasta que lo terminamos ahi paso a ser falso entonces se terminaria만l막ucle

Vamos a aplicar nuestro conocimiento reci칠n adquirido sobre bucles para resolver un problema real: calcular el promedio de las calificaciones de un estudiante. Supongamos que un estudiante nos proporciona sus calificaciones una por una. El estudiante indicar치 que ha terminado de proporcionar las calificaciones ingresando el n칰mero 0.

```python
suma_calificaciones = 0
contador_asignaturas = 0

calificacion = int(input("Ingrese una calificaci칩n (entre 1 y 12) o 0 para terminar: "))

while calificacion != 0:
    suma_calificaciones += calificacion
    contador_asignaturas += 1
    calificacion = int(input("Ingrese una calificaci칩n (entre 1 y 12) o 0 para terminar: "))

promedio = suma_calificaciones / contador_asignaturas
print(f"El promedio de las calificaciones es: {promedio}")
```

En este c칩digo, el bucle while se ejecuta hasta que el estudiante ingrese 0. Si bien es sencillo y cumple con su objetivo, este c칩digo no verifica si las calificaciones ingresadas est치n dentro del rango v치lido (1 a 12).

## d. Controlando el Flujo con 'break' y 'continue'

Ahora, vamos a mejorar nuestro c칩digo anterior incorporando las declaraciones break y continue para un mejor control del flujo del programa y la validaci칩n de las calificaciones.

En este c칩digo, la declaraci칩n break se utiliza para detener el bucle cuando el estudiante ingresa 0. Por otro lado, la declaraci칩n continue se usa para saltar a la pr칩xima iteraci칩n del bucle si la calificaci칩n ingresada no est치 dentro del rango v치lido, evitando as칤 que calificaciones incorrectas se sumen a suma_calificaciones y se cuenten en contador_asignaturas.

```python
suma_calificaciones = 0
contador_asignaturas = 0

while True:
    calificacion = int(input("Ingrese una calificaci칩n (entre 1 y 12) o 0 para terminar: "))
    
    if calificacion == 0:
        break
    
    if calificacion < 1 or calificacion > 12:
        print("La calificaci칩n debe estar entre 1 y 12. Int칠ntalo de nuevo.")
        continue

    suma_calificaciones += calificacion
    contador_asignaturas += 1

promedio = suma_calificaciones / contador_asignaturas
print(f"El promedio de las calificaciones es: {promedio}")
```

## Desaf칤o 3: Simulaci칩n de una Carrera de Autos

Vas a simular una carrera de autos. Cada auto tiene una velocidad aleatoria (puedes usar la biblioteca random de Python) y cada ciclo del bucle representa un segundo de la carrera. Al final de cada segundo, cada auto avanza una distancia igual a su velocidad. La carrera dura 10 segundos. Al final de la carrera, debes imprimir el auto ganador. Si hay un empate, debes imprimir todos los autos que empataron.

*Nota:* Este desaf칤o puede requerir que aprendas sobre conceptos adicionales, por ejemplo c칩mo generar n칰meros aleatorios.

```python
import random

# Lista de autos participantes
autos = ['Auto 1', 'Auto 2', 'Auto 3', 'Auto 4']

# Diccionario para almacenar la distancia recorrida por cada auto
distancias = {auto: 0 for auto in autos}

# Simular la carrera por 10 segundos
for segundo in range(1, 11):
    print(f"Segundo {segundo}:")
    for auto in autos:
        velocidad = random.randint(1, 10)  # velocidad aleatoria entre 1 y 10 m/s
        distancias[auto] += velocidad
        print(f"  {auto} avanza {velocidad} m (Total: {distancias[auto]} m)")
    print()  # l칤nea en blanco para separar segundos

# Determinar la m치xima distancia recorrida
max_distancia = max(distancias.values())

# Determinar el/los ganador(es)
ganadores = [auto for auto, distancia in distancias.items() if distancia == max_distancia]

# Mostrar resultados
print("游끠 Resultado final:")
for auto, distancia in distancias.items():
    print(f"{auto}: {distancia} metros")

print("\n游볞 Ganador(es):")
print(", ".join(ganadores))
```

## Problema: C치lculo de Gastos de un Viaje

Supongamos que est치s planeando un viaje y quieres calcular el total de gastos estimados. Tienes un arreglo de los diferentes conceptos de gastos (por ejemplo, transporte, alojamiento, alimentaci칩n, actividades, etc.) y quieres solicitar al usuario que ingrese el costo estimado para cada concepto. Al final, quieres mostrar el total de gastos estimados para el viaje.

Planteamiento del problema:
* Crea un arreglo de conceptos de gastos.
* Utiliza un bucle for para iterar sobre el arreglo y solicitar al usuario que ingrese el costo estimado para cada concepto.
* Utiliza una variable para realizar el seguimiento del total de gastos estimados, inicializada en 0.
* Dentro del bucle, agrega el costo ingresado a la variable del total de gastos estimados.
* Al final del bucle, muestra el total de gastos estimados.

Aqu칤 tienes un ejemplo de c칩mo podr칤as resolver este desaf칤o utilizando arreglos:

```python
import numpy as np

conceptos_gastos = np.array(['Transporte', 'Alojamiento', 'Alimentaci칩n', 'Actividades', 'Otros'])

total_gastos = 0

for concepto in conceptos_gastos:
    costo = float(input(f"Ingrese el costo estimado para {concepto}: "))
    total_gastos += costo

print(f"El total de gastos estimados para el viaje es: {total_gastos}")
```

En este c칩digo, utilizamos un bucle for para iterar sobre el arreglo de conceptos de gastos. En cada iteraci칩n, solicitamos al usuario que ingrese el costo estimado para ese concepto y lo sumamos al total de gastos estimados. Al final, mostramos el resultado.

## Problema: An치lisis de Calificaciones utilizando Pandas y Bucles 
Supongamos que tienes un archivo CSV llamado "calificaciones.csv" que contiene las calificaciones de varios estudiantes. Tu objetivo es realizar un an치lisis de las calificaciones utilizando la biblioteca Pandas y bucles for.


*Planteamiento del problema:*
Importa la biblioteca Pandas.
Carga el archivo CSV "calificaciones.csv" en un DataFrame de Pandas.
Utiliza un bucle for para calcular el promedio de cada estudiante y mostrarlo en la consola.

Aqu칤 tienes un ejemplo de c칩mo podr칤as resolver este desaf칤o utilizando Pandas y bucles for:

```python
import pandas as pd

# Carga el archivo CSV en un DataFrame
df = pd.read_csv("calificaciones.csv")

# Inicializa variables para calcular el promedio general
total_calificaciones = 0
numero_calificaciones = 0

# Utiliza un bucle for para sumar todas las calificaciones
for index, row in df.iterrows():
    nombre = row['Nombre']
    calificacion = int(row['Calificacion'])
    total_calificaciones += calificacion
    numero_calificaciones += 1
    print(f"La calificaci칩n de {nombre} es: {calificacion}")

# Calcula el promedio general
promedio_general = total_calificaciones / numero_calificaciones
print(f"El promedio general de las calificaciones es: {promedio_general:.2f}")
```

En este c칩digo, utilizamos Pandas para cargar el archivo CSV en un DataFrame. Luego, utilizamos un bucle for junto con iterrows() para iterar sobre cada fila del DataFrame. Para cada estudiante, extraemos el nombre y la calificaci칩n, para luego calcular el promedio.
